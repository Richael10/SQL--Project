```python
import numpy as np
import scipy.stats as stats

# Generate sample data (replace this with your actual data)
np.random.seed(0)  # for reproducibility
sample_data = np.random.normal(loc=100, scale=20, size=100)  # example data with mean 100 and standard deviation 20

# Calculate the confidence interval
confidence_level = 0.95  # desired confidence level (e.g., 95%)
degrees_freedom = len(sample_data) - 1  # degrees of freedom for the t-distribution
sample_mean = np.mean(sample_data)
sample_standard_error = stats.sem(sample_data)
confidence_interval = stats.t.interval(confidence_level, degrees_freedom, loc=sample_mean, scale=sample_standard_error)

# Print the confidence interval
print(f"Confidence Interval ({confidence_level * 100}%): {confidence_interval}")
```

```python
import numpy as np

# Generate fake data
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=100, scale=20, size=100)  # Mean=100, Std Dev=20, Sample size=100

# Calculate mean and standard deviation of the data
mean = np.mean(data)
std_dev = np.std(data)

# Define confidence level (e.g., 95% confidence level)
confidence_level = 0.95

# Calculate the standard error (SE) using the formula SE = std_dev / sqrt(sample_size)
sample_size = len(data)
standard_error = std_dev / np.sqrt(sample_size)

# Calculate the margin of error (MOE) using the formula MOE = critical_value * SE
# For a 95% confidence level, the critical value is approximately 1.96 (for a large sample size)
critical_value = 1.96
margin_of_error = critical_value * standard_error

# Calculate the confidence interval using the formula CI = (mean - MOE, mean + MOE)
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

print(f"Mean: {mean}")
print(f"Standard Deviation: {std_dev}")
print(f"Confidence Interval ({confidence_level*100}%): {confidence_interval}")
```

```python
import pandas as pd
import numpy as np

# Creating fake data
np.random.seed(0)
data = {
    'group': np.random.choice(['A', 'B', 'C'], size=100),
    'value': np.random.randint(1, 100, size=100)
}
df = pd.DataFrame(data)

# Calculate group means
group_means = df.groupby('group')['value'].mean()

# Calculate total mean
total_mean = df['value'].mean()

# Calculate sum of squares between groups
ss_between = ((group_means - total_mean) ** 2).sum() * len(group_means)

# Calculate sum of squares within groups
ss_within = df.groupby('group')['value'].apply(lambda x: ((x - x.mean()) ** 2).sum()).sum()

# Calculate degrees of freedom
df_between = len(group_means) - 1
df_within = len(df) - len(group_means)

# Calculate mean squares
ms_between = ss_between / df_between
ms_within = ss_within / df_within

# Calculate F-statistic
f_statistic = ms_between / ms_within

# Calculate p-value
from scipy.stats import f as f_dist
p_value = 1 - f_dist.cdf(f_statistic, df_between, df_within)

# Print results
print(f'F-Statistic: {f_statistic:.2f}')
print(f'P-Value: {p_value:.4f}')
```

```python
# Install required packages (run this if not already installed)
# !pip install numpy pandas statsmodels scipy

# Import necessary libraries
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import f

# Create the fake dataset
np.random.seed(42)

# Sample data for three groups
group1 = np.random.normal(loc=20, scale=5, size=30)
group2 = np.random.normal(loc=22, scale=5, size=30)
group3 = np.random.normal(loc=19, scale=5, size=30)

# Creating a DataFrame
data = pd.DataFrame({
    'value': np.concatenate([group1, group2, group3]),
    'group': ['group1']*30 + ['group2']*30 + ['group3']*30
})

# Displaying the first few rows of the dataset
print(data.head())

# Perform the ANOVA test
# Fit the model
model = ols('value ~ C(group)', data=data).fit()

# Perform ANOVA
anova_table = sm.stats.anova_lm(model, typ=2)

# Display the ANOVA table
print(anova_table)

# Calculating SSW, SSB, SST, F-statistic, and p-value manually
# Mean of each group
mean_group1 = np.mean(group1)
mean_group2 = np.mean(group2)
mean_group3 = np.mean(group3)

# Overall mean
overall_mean = np.mean(data['value'])

# Sum of Squares Within (SSW) - also known as Residual Sum of Squares (SSR)
ssw = np.sum((group1 - mean_group1) ** 2) + np.sum((group2 - mean_group2) ** 2) + np.sum((group3 - mean_group3) ** 2)

# Sum of Squares Between (SSB) - also known as Sum of Squares Model (SSM)
ssb = (len(group1) * (mean_group1 - overall_mean) ** 2) + (len(group2) * (mean_group2 - overall_mean) ** 2) + (len(group3) * (mean_group3 - overall_mean) ** 2)

# Sum of Squares Total (SST)
sst = ssw + ssb

# Degrees of freedom
df_between = 2  # Number of groups - 1
df_within = len(data) - 3  # Total number of observations - Number of groups

# Mean Squares
ms_between = ssb / df_between
ms_within = ssw / df_within

# F-statistic
f_statistic = ms_between / ms_within

# p-value
p_value = 1 - f.cdf(f_statistic, df_between, df_within)

print("\nManual Calculation:")
print(f"SSW (Within): {ssw:.2f}")
print(f"SSB (Between): {ssb:.2f}")
print(f"SST (Total): {sst:.2f}")
print(f"F-statistic: {f_statistic:.2f}")
print(f"p-value: {p_value:.4f}")
```

```python
import scipy.stats as stats

# Standardize column names
df.columns = df.columns.str.strip().str.lower()

# Check if necessary columns exist
if "salary" in df.columns and "department" in df.columns:
    # Drop missing values
    df = df.dropna(subset=["salary", "department"])

    # Convert Salary to numeric
    df["salary"] = pd.to_numeric(df["salary"], errors="coerce")

    # Ensure at least two departments exist
    if df["department"].nunique() > 1:
        # Group salaries by department
        groups = [group["salary"].dropna() for _, group in df.groupby("department")]

        # Perform ANOVA test
        anova_result = stats.f_oneway(*groups)

        # Print results
        print(f"ANOVA F-statistic: {anova_result.statistic:.2f}, P-value: {anova_result.pvalue:.4f}")

        # Interpretation
        if anova_result.pvalue < 0.05:
            print("There is a significant difference in salary across departments.")
        else:
            print("No significant difference in salary across departments.")
    else:
        print("Not enough department data for ANOVA.")
else:
    print("Error: 'salary' or 'department' column not found. Please check column names.")
```
